```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
df= pd.read_csv("adver.csv")
X= df.drop(['Sales', 'Unnamed: 0'], axis=1)
Y=df['Sales'].values.reshape(-1,1)
print(df.head())
print(df.shape)
# X contains the three columns Tv radio and newspaper
#  Y contains only sales in matrix of n rows and 1 column
# print(X)
# print(Y)
```

```python
x_train, x_test, y_train, y_test= train_test_split(X, Y, test_size=1/3, random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
reg= LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_pred= reg.predict(x_test)
print(y_pred)
```

```python
sb.pairplot(df, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', height=7, aspect=0.8, kind='reg')
```

```python
mse=mean_squared_error(y_test,y_pred)
rmse=(mse)**(1/2)
mae=mean_absolute_error(y_test,y_pred)
rsq = r2_score(y_test,y_pred)
print("MSE : ", mse)
print("MAE :", mae)
print("RSQ : ",rsq)
```
