```python
import pandas as pd
df= pd.read_csv('adver.csv')
df.head()
```

```python
df.drop(['Unnamed: 0', 'TV' ,'Sales'], inplace = True, axis=1)
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10,7))
plt.title("Dendrogram - AGNES")
dend= shc.dendrogram(shc.linkage(df, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster= AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage ='ward')
cluster.fit_predict(df)
```

```python
cl=cluster.fit_predict(df)
X=df.values
plt.figure(figsize=(10,7))
plt.scatter(X[cl==0, 0], X[cl==0, 1], s=100, c='orange', label='Cluster 1')
plt.scatter(X[cl==1, 0], X[cl==1, 1], s=100, c='blue', label='Cluster 1')
plt.title('Sales')
plt.xlabel('Radio')
plt.ylabel('Newspaper')
plt.show()
```

```python
print("DIANA")
df= pd.read_csv('adver.csv')
df.head()
```

```python
df.drop(['Unnamed: 0', 'TV' ,'Sales'], inplace = True, axis=1)
df.head()
```

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrogram - DIANA")
dend= shc.dendrogram(shc.linkage(df, method='ward'), orientation='bottom', distance_sort='ascending')
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering_model= AgglomerativeClustering(n_clusters=50, affinity='euclidean', linkage='ward')
clustering_model.fit(df)
clustering_model.fit_predict(df)
```

```python
import seaborn as sb
data_labels= clustering_model.labels_
sb.scatterplot(x='Radio', y='Newspaper', data=df, hue=data_labels, palette= 'rainbow')
plt.title("DIANA")
plt.show()
```
